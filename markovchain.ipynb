```python
import pandas as pd
import numpy as np
import markovify
```

```python
df = pd.read_csv("chords.csv")
df.tune = df.tune.apply(lambda x: eval(x))
df
```

```python
train = df.tune.to_numpy().tolist()
train
```

```python
model = markovify.Chain(train, state_size=8)
```

```python
for c in model.gen():
    print(c)
```
